```python
import numpy as np
import pandas as pd
import torch
#!pip install transformers
import transformers as ppb # pytorch transformers
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.model_selection import StratifiedKFold

torch.cuda.device_count()
```

```python
df = pd.read_csv('haha_2019_train.csv')
```

```python
df.head()
```

```python
pretrained_weights = 'distilroberta-base'

from transformers import AutoTokenizer, AutoModelForMaskedLM
  
tokenizer_class = AutoTokenizer.from_pretrained("distilroberta-base",add_prefix_space=True)

model_class = AutoModelForMaskedLM.from_pretrained("distilroberta-base")

tokenizer = tokenizer_class.from_pretrained(pretrained_weights)
model = model_class.from_pretrained(pretrained_weights)
```

```python
class1 = df.loc[df["is_humor"]==1][:300]
class0 = df.loc[df["is_humor"]==0][:300]
result = class1.append(class0, sort=False)
tokenized = result["text"].apply((lambda x: tokenizer.encode(x, add_special_tokens=True)))
```

```python
max_len = 0
for i in tokenized.values:
    if len(i) > max_len:
        max_len = len(i)

padded = np.array([i + [0]*(max_len-len(i)) for i in tokenized.values])

attention_mask = np.where(padded != 0, 1, 0)
attention_mask.shape
```

```python
cuda0 = torch.device('cuda:0')

input_ids = torch.tensor(padded)  
attention_mask = torch.tensor(attention_mask)

with torch.no_grad():
    last_hidden_states = model(input_ids, attention_mask=attention_mask)
```

```python
features = last_hidden_states[0][:,0,:].numpy()
labels = result["is_humor"]
```

```python
skf = StratifiedKFold(n_splits=10, random_state=0, shuffle=True)
for train_index, test_index in skf.split(features,labels):
  train_features = features[train_index]
  train_labels = result.iloc[train_index]["is_humor"]

  test_features = features[test_index]
  test_labels = result.iloc[test_index]["is_humor"]

  lr_clf = LogisticRegression(C=10)
  lr_clf.fit(train_features, train_labels)

  print(classification_report(lr_clf.predict(test_features),test_labels))
```

```python
skf = StratifiedKFold(n_splits=10, random_state=0, shuffle=True)
for train_index, test_index in skf.split(features,labels):
  train_features = features[train_index]
  train_labels = result.iloc[train_index]["is_humor"]

  test_features = features[test_index]
  test_labels = result.iloc[test_index]["is_humor"]

  lr_clf = LogisticRegression(C=10)
  lr_clf.fit(train_features, train_labels)

  print(classification_report(lr_clf.predict(test_features),test_labels))
```
